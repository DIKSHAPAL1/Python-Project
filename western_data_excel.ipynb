```python
#The error message indicates that there are non-finite values (NaN or inf) present in the 'Gross Sales' column, making it impossible to convert the column to integer type directly. To resolve this issue, you can either remove or replace the non-finite values before converting the column to integer type.

#Here's how you can do it:

#Replace NaN values: You can replace NaN values with a default value (e.g., 0) before converting the column to integer type.#
#Remove non-finite values: You can remove rows containing non-finite values from the DataFrame before converting the column to integer type.#
#NaN values are replaced with 0 using the fillna() method.
#Rows containing non-finite values are removed using boolean indexing with notna() and apply().
#Finally, the 'Gross Sales' column is converted to integer type using astype(int)
```

```python
import pandas as pd

# Replace NaN values with 0
western_countries['Profit'].fillna(0, inplace=True)

# Remove rows containing non-finite values (NaN or inf)
western_countries = western_countries[western_countries['Profit'].notna() & western_countries['Profit'].apply(lambda x: not pd.isnull(x) and pd.notna(x) and x != float('inf'))]

# Convert 'Profit' column to integer type
western_countries['Profit'] = western_countries['Profit'].astype(int)

print(western_countries['Profit'].dtype)
```

```python
western_countries['Units_Sold'].fillna(0, inplace=True)
western_countries = western_countries[western_countries['Units_Sold'].notna() & western_countries['Units_Sold'].apply(lambda x: not pd.isnull(x) and pd.notna(x) and x != float('inf'))]
western_countries['Units_Sold'] = western_countries['Units_Sold'].astype(int)
print(western_countries['Units_Sold'].dtype)
```
