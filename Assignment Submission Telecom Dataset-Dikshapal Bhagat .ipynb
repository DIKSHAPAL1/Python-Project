#Dataset Columns Information**

PID = a unique identifier for the phone model

Blue = whether the phone has bluetooth support or not

Wi_Fi = whether the phone has wifi support or not

Tch_Scr = whether the phone has touch screen support or not

Ext_Mem = whether the phone has external memory support or not

Px_h = number of pixels in the vertical axis of the phone

Px_w = number of pixels in the horizontal axis of the phone

Scr_h = height of the screen of the phone in centimetres (cm)

Scr_w = width of the screen of the phone in centimetres (cm)

Int_Mem = internal memory of the phone measured in megabytes (MB)

Bty_Pwr = maximum energy stored by the phone's battery measured in 
milli-Ampere-hours (mAh)

PC = resolution of the primary camera measued in megapixels (MP)

FC = resolution of the front camera measued in megapixels (MP)

RAM = random access memory available in the phone measured in gigabytes (GB)

Depth = depth of the mobile phone measured in centimetres (cm)

Weight = weight of the mobile phone measured in grams (g)

Price = selling price of the mobile phone in rupees

```python
# Screen Resolution 
file['Screen_Resolution'] = file['Scr_h'] * file['Scr_w']
print("Screen Resolution Summary:")
print(file['Screen_Resolution'].describe())


median_resolution = file['Screen_Resolution'].median()
con2 = file['Screen_Resolution'] >= median_resolution

# Logical Values 
print("\nLogical Condition (con2):")
print(con2)

filtered_phones = file[con2]
print("\nPhones with Screen Resolution >= Median:")
print(filtered_phones)
```

```python


# Estimate Scr_d By using pythogorus theorem
file['Scr_d'] = np.sqrt(file['Scr_h']**2 + file['Scr_w']**2)

# Scr_d 
print("Screen Diagonal (Scr_d) Summary:")
print(file['Scr_d'].describe())

print("\nUpdated DataFrame with Scr_d:")
print(file[['Scr_h', 'Scr_w', 'Scr_d']])
```

```python

# Estimate quartile of Scr_d features
q1 = file['Scr_d'].quantile(0.25)  
q2 = file['Scr_d'].quantile(0.50) 
q3 = file['Scr_d'].quantile(0.75)  

# Histogram 
plt.figure(figsize=(10, 6))
plt.hist(file['Scr_d'], bins=15, color='skyblue', edgecolor='black', alpha=0.7)

#Add Quartile line
plt.axvline(x=q1, color='red', linestyle='--', label=f"Q1 (25%): {q1:.2f}")
plt.axvline(x=q2, color='green', linestyle='--', label=f"Median (Q2 - 50%): {q2:.2f}")
plt.axvline(x=q3, color='blue', linestyle='--', label=f"Q3 (75%): {q3:.2f}")

plt.title('Histogram of Screen Diagonal (Scr_d)', fontsize=16)
plt.xlabel('Screen Diagonal (Scr_d)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()


plt.show()
```

```python
# Step 1: Calculate the upper quartile (Q3) value of the screen size
upper_quartile_screen_size = file['Scr_d'].quantile(0.75)

# Step 2: Create the logical condition for very good screen sizes
con3 = file['Scr_d'] >= upper_quartile_screen_size

# Step 3: Display the condition and filtered phones
print("Logical Condition (con3):")
print(con3)
```

```python
# filter good screen resolution of phones
very_good_screen_phones = file[con3]
print("\nPhones with Very Good Screen Sizes (>= Upper Quartile):")
print(very_good_screen_phones[['Scr_d']])
```
